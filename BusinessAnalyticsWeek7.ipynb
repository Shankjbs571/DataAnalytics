```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np

data = {
    'Sales': [53, 12, 15, 27, 8, 12, 27, 17, 61, 20, 95, 33, 19, 15, 43, 11, 20, 36, 7, 9,
              21, 25, 29, 18, 6, 7, 15, 23, 51, 93, 51, 14, 41, 8, 114, 35, 18, 9, 46, 17,
              43, 63, 8, 34, 145, 18, 40, 17, 17, 12],
    'Price': [1, 3.4, 3.3, 1.4, 3.7, 3.6, 2.2, 1.7, 1.4, 3.2, 0.8, 1.1, 2.5, 3.3, 1, 2.6, 3,
              2.4, 3.6, 2.8, 2.4, 2.3, 2.1, 1.8, 3.6, 3.7, 2.9, 1.2, 1.1, 1.2, 1, 3.7, 1.2,
              3.4, 1, 2.5, 2, 3.3, 1.8, 2, 1.3, 0.8, 3.6, 1.6, 0.9, 2.4, 1.7, 2.8, 3.2, 3.7]
}

df = pd.DataFrame(data)

model = smf.ols('Sales ~ Price', data=df).fit()

r_squared = model.rsquared

print(f'R-squared value: {r_squared:.2f}')
```

```python
print(model.summary())
```

```python

df['log_Sales'] = np.log(df['Sales'])
df['log_Price'] = np.log(df['Price'])

model_log = smf.ols('log_Sales ~ log_Price', data=df).fit()

print(model_log.summary())

r_squared_log = model_log.rsquared
print(f"R-squared value on the transformed dataset: {r_squared_log:.2f}")
```
